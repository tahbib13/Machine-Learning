```python
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image, display_svg, SVG
```

```python
clas = tree.DecisionTreeClassifier()
iris = load_iris()

X_train = iris.data
y_train = iris.target
clas.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(clas)
```

# 1. Classification


```python
import dtreeviz

# Assuming 'clas', 'X_train', 'y_train', 'iris.feature_names',
# and the class names are already defined and prepared as in previous steps.

viz = dtreeviz.model(clas,
                     X_train=X_train, # Ensure X_train is passed with keyword argument
                     y_train=y_train, # Ensure y_train is passed with keyword argument
                     feature_names=iris.feature_names,
                     target_name='species', # Add a target_name
                     class_names=["setosa", "versicolor", "virginica"])

# To display the visualization
viz.view()
```

# 2. Regression


```python
import dtreeviz
from sklearn import tree
from sklearn.datasets import load_diabetes

# For example with load_diabetes:
diabetes = load_diabetes()
X_train = diabetes.data
y_train = diabetes.target
regr = tree.DecisionTreeRegressor(max_depth=3)
regr.fit(X_train, y_train)

# Corrected: Removed the 'scale' argument
viz = dtreeviz.model(regr,
                     X_train=X_train,
                     y_train=y_train,
                     target_name='disease_progression',
                     feature_names=diabetes.feature_names
                     # Removed: scale=2
                    )

viz.view()
```

# 3. Horizontal Decision Tree


```python
import dtreeviz
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_diabetes # Using load_diabetes as a regression example

# 1. Load a regression dataset (e.g., Diabetes)
diabetes = load_diabetes()
X_train = diabetes.data
y_train = diabetes.target

# 2. Train a Decision Tree REGRESSOR
clas = DecisionTreeRegressor(max_depth=3) # Assuming 'clas' is your regressor
clas.fit(X_train, y_train)

# 3. Correct dtreeviz.model() call for REGRESSION
#    - Use dtreeviz.model() instead of dtreeviz()
#    - Remove 'class_names' as it's a regressor
#    - 'scale' and 'orientation' are *not* arguments for dtreeviz.model(). They are for the .view() method.
viz = dtreeviz.model(clas,
                     X_train=X_train,
                     y_train=y_train,
                     target_name='disease_progression', # Appropriate for diabetes
                     feature_names=diabetes.feature_names
                    )

# 4. Display the visualization using .view()
#    - Pass 'scale' and 'orientation' to the .view() method, not .model()
viz.view(scale=1.5, orientation='LR') # Pass scale and orientation here!
```
